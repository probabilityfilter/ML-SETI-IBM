#### Read the Data Set

```python
len(raw_data)
```

### Use `ibmseti` to analyze the data


```python
aca = ibmseti.compamp.Compamp(raw_data) #archive-compamp (ACA)
spectrogram = ibmseti.dsp.compamp_to_spectrogram(aca)  
#calculates the frequency-bin and time-bin edges
freqbins = ibmseti.dsp.frequency_bins(aca.header())
timebins = ibmseti.dsp.time_bins(aca.header())
```

```python
aca.header() #here's what the data header looks like, if you're curious. It probably won't be very useful in analysis.
```

```python
print list(first_row['signaldb_rows'])  #just printing this to show it's all there.
```

```python
fig, ax = plt.subplots()
ax.pcolormesh(freqbins, timebins, spectrogram)  #Notice this is a "waterfall" plot going up. Typically t=0 is at the top of a "waterfall" plot
```

```python
spectrogram.max()
```

```python
spectrogram.mean()
```

## More Spacecraft Data

You can get all of the spacecraft data from this endpoint [/v1/meta/aca/spacecraft](http://setigopublic.mybluemix.net/v1/aca/meta/spacecraft). (REST API documentation: setigopublic.md)
The JSON data returned from this URL is the same as returned from the `/v1/meta/aca` endpoint. Each row will contain the meta-data for an observed signal from some type of spacecraft, including the container and objectname of the raw data file, which you can use to download the data and analyze.
